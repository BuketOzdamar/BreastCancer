```python
#Kütüphaneler;
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Data oku
data_csv = r"data.csv"
data = pd.read_csv(data_csv)
data.head()
```

| Kolon Adı                 | Açıklama                                                        |
|---------------------------|----------------------------------------------------------------|
| Id                        | Örnek hasta numarası / benzersiz kimlik                        |
| ClumpThickness            | Tümörün kümelenme kalınlığı (1-10 arası değer)                 |
| UniformityCellSize        | Hücre boyutlarının uniformluğu (homojenlik)                    |
| UniformityCellShape       | Hücre şekillerinin uniformluğu                                 |
| MarginalAdhesion          | Hücrelerin kenar yapışkanlığı                                  |
| SingleEpithelialCellSize  | Tek tek epitel hücrelerinin boyutu                             |
| BareNuclei                | Çıplak çekirdek sayısı veya yoğunluğu                         |
| BlandChromatin            | Kromatin dokusunun yumuşaklığı                                |
| NormalNucleoli            | Normal çekirdekçik sayısı                                     |
| Mitoses                   | Bölünme sayısı (mitoz)                                        |
| Malignite                 | Malignite durumu (2 = benign, 4 = malign) ya da sınıflandırma |

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = data.drop(columns=['Id'])

# Korelasyon matrisini oluştur
corr_matrix = df.corr()

# Korelasyon matrisini görselleştir
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Özellikler Arası Korelasyon Matrisi", fontsize=14)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
features = ['ClumpThickness ', 'UniformityCellSize', 'UniformityCellShape', 
            'MarginalAdhesion', 'SingleEpithelialCellSize']

for feature in features:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x='Malignite', y=feature, data=df, palette='coolwarm')
    plt.title(f"{feature} - Malignite Sınıflarına Göre Dağılım")
    plt.show()
```

```python
sns.pairplot(df[['ClumpThickness ', 'UniformityCellSize', 'UniformityCellShape', 'Malignite']], 
             hue='Malignite', palette='husl')
plt.suptitle("Bazı Özellikler Arası İlişkiler (Pairplot)", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='UniformityCellSize', hue='Malignite', kde=True, multiple="stack", palette='Set1')
plt.title("UniformityCellSize Dağılımı - Malignite Sınıfına Göre")
plt.show()
```

```python
sns.violinplot(x='Malignite', y='SingleEpithelialCellSize', data=df, palette='muted')
plt.title("SingleEpithelialCellSize - Malignite'e Göre Violin Plot")
plt.show()
```

```python
sns.countplot(x='Malignite', data=df, palette='Set2')
plt.title("Malignite Sınıf Dağılımı")
plt.show()

```
